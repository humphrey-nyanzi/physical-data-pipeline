```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
updf_data = pd.read_csv('./13_Feb_UPDF.csv').reset_index()
kitara_data = pd.read_csv('./13_Feb_Kitara.csv').reset_index()
```

```python
match_data = pd.concat([updf_data,kitara_data]).reset_index()
```

```python
match_data['Team'] = match_data['Team'].fillna('').str.strip()
match_data['Team'] = match_data['Team'].apply(lambda x: 'UPDF' if 'UPDF' in x or 'UPF' in x else 'Kitara')

match_data['Player Name'] = match_data['Player Name'].fillna('').str.strip().str.title()
match_data['Position'] = match_data['Position'].str.strip().str.upper()
match_data['Position'] = match_data['Position'].apply(lambda x: 'FW' if 'FWD' in x else x )


match_data.drop(match_data.columns[23:],axis='columns',inplace=True)
match_data.drop(['Date','Split Start Time','Split End Time','Hr Load','Time In Red Zone (min)','index','Hr Max (bpm)'],axis='columns',inplace=True)
match_data.columns = match_data.columns.str.lower().str.replace(' ', '_')
```

```python
match_data.drop('level_0',axis='columns',inplace=True);
```

```python
match_data.shape
match_data.describe()
```

```python
corr_df = match_data.corr(numeric_only=True)
sns.heatmap(corr_df);
```

```python
# Select only numeric columns
numeric_cols = match_data.select_dtypes(include=[np.number]).columns

# Group by position and aggregate by mean
position_grouped = match_data.groupby('position')[numeric_cols].mean().reset_index()

# Group by team and aggregate by mean
team_grouped = match_data.groupby('team')[numeric_cols].mean().reset_index()
```
